```python
import pandas as pd
import numpy as np
```

```python
# load data
data = pd.read_excel("C:/Users/ElohorEfere/Documents/DATA ANALYTICS/PYTHON/PANDAS PRACTISE DATASET.xlsx")
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:")
missing_values
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Deal with missing values in Date Column
data['Date'] = data['Date'].ffill()

```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:")
missing_values
```

```python
# Deal with missing values in Calories Column
imputer = SimpleImputer(strategy = 'median')
imputer.fit(data[['Calories']])
data[['Calories']] = imputer.transform(data[['Calories']])
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:")
missing_values
```

```python
# Check for Data Types
data.dtypes
```

```python
# Reassign Data Types
data['Date'] = data['Date'].astype('datetime64[ns]')
```

```python
# Check for Data Types
data.dtypes
```

```python
# Check for Duplicates
duplicates = data.duplicated()
data[duplicates]
```

```python
# Remove Duplicates
data.drop_duplicates(inplace = True)
```
